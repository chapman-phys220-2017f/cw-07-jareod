 Below is the basic Taylor series approximation formula:

$$f(a + y) = \sum_{n=0}^\infty \frac{y^n}{n!}\left[\frac{d^n}{dy^n}f(a+y)\right]_{y=0} = f(a) + f'(a)y + f''(a)\frac{y^2}{2} + \cdots$$



The taylor series expansion approximation represents a function that uses an infinite sum of terms that come from its derivatives around a given point.  The left side of the equation is describing the summation of the derivatives using $\Sigma$ from $n = $ 0 to $\infty$.  The right side of the equation is demonstrating the summation term by term, or the left side of the equation being executed.

The expansion does not have to be infinite, in-fact many times has a finite number of terms.  In turn, this expansion is basically a limit of the number of expansions made within the bounds of the sum.  While the series converges to a point, it may not be the exact value of the function.  This is why it is an approximation, but the error between the exact value and the approximated value tends to be quite small.

$$f(x) = \sum_{n=0}^\infty \frac{(x-a)^n}{n!}\left[\frac{d^n}{dx^n}f(x)\right]_{x=a} = f(a) + f'(a)(x-a) + f''(a)\frac{(x-a)^2}{2} + \cdots$$

This formulation makes it more clear that the function $f$ at any point $x$ in the domain may be approximated by derivatives at $a$ and the displacements $(x-a)$ away from the point .
